```python
import tvkit
import asyncio
from tvkit.api.chart.ohlcv import OHLCV

while True:
    async with OHLCV() as client:
        count = 0
        async for bar in client.get_ohlcv("NSE:NIFTY", interval="1"):
            count += 1
            print(f"Bar {count}: NIFTY ${bar.close:,.2f} | Volume: {bar.volume:,.0f}")
    if count >= 50:
        break
```
